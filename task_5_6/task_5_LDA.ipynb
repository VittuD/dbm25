```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.preprocessing import MinMaxScaler
import os
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
feature_space = torch.load("/workspaces/dbm25/data/extracted_features.pt")
feature_type = "layer3"
k = 5  # Numero di topic

X = []
image_ids = []

for element in feature_space:
    if feature_type in element:
        X.append(element[feature_type])
        path = element['file_path']
        img_id = os.path.basename(path)
        image_ids.append(img_id)

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

lda = LatentDirichletAllocation(n_components=k, random_state=0)
X_lda = lda.fit_transform(X_scaled)  

lda_results = {
    img_id: topic_weights.tolist()
    for img_id, topic_weights in zip(image_ids, X_lda)
}

# Salva in un file .pt
torch.save(lda_results, "/workspaces/dbm25/data/lda_image_topics.pt")
print(lda_results)
print(f"Shape of LDA results: {len(lda_results)}, {len(next(iter(lda_results.values())))}")
```
